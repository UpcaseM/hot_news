## 数据源获取

这里借用了[dataabc](https://github.com/dataabc)写的微博评论爬虫。我爬取了4月23－4月25的部分**原创微博**。

数据源分为２个微博主题：１.罗志祥周扬青分手。　２.罗志祥回应

我们先看下主题一各位都给出怎么样的评价，然后对于罗志祥的回应主题大家又有什么反映。

```python
import os
import re
import jieba
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from pylab import mpl
import seaborn as sns
from itertools import cycle
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
pd.set_option('max_columns', 50)
plt.style.use('bmh')
color_cycle = cycle(plt.rcParams['axes.prop_cycle'].by_key()['color'])
mpl.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题

project_dir = '/mnt/data-ubuntu/Projects/hot_news_analysis'
```

```python
df_fs = pd.read_csv('/mnt/data-ubuntu/Projects/hot_news_analysis/data/罗志祥分手20200423/罗志祥周扬青分手.csv')
df_hy = pd.read_csv('/mnt/data-ubuntu/Projects/hot_news_analysis/data/罗志祥分手20200423/罗志祥回应.csv')
```

```python
df_fs = df_fs.loc[:,['id', '用户昵称', '微博正文', '发布位置', '话题', '转发数', '评论数', '点赞数', '发布时间', '发布工具']]
df_fs.columns = ['id', 'name', 'text', 'location', 'topic', 'num_share', 'num_comm', 'num_like', 'time_stamp', 'platform']
df_fs['time_stamp'] = pd.to_datetime(df_fs['time_stamp'])
df_fs['date'] = df_fs['time_stamp'].dt.date
df_fs['hour'] = df_fs['time_stamp'].dt.hour
df_fs
```

```python
def generate_wd(text, stop_words, save_to = ''):
    """
    Function to generate word cloud.
    
    Param
        text: string
            a string contains words to create world cloud.
        stop_words: list
            list of words to seprate the text.
        save_to: string
            path to save the wordclound graph.
    Return
        None
    """
    wc = WordCloud(background_color='white',
               max_words=300,
               #mask=back_color,
               max_font_size=100,
               stopwords=list(STOPWORDS) + stop_words,
               font_path='/home/rick/.local/share/fonts/SimHei.ttf',
               random_state=42,
               width=600,
               height=200
               )
    wc.generate(text)
    fig, ax1 = plt.subplots(figsize=(20, 6))
    plt.imshow(wc)
    plt.axis('off')
    if save_to != '':
        wc.to_file(save_to)
```

```python
stop_words = ['罗志祥周扬青分手', '罗志祥', '小猪', '周扬青', '微博', '视频', '网页','链接']
for w in stop_words:
    jieba.add_word(w)
df_fs['text_split'] = df_fs['text'] \
    .map(lambda x: \
         ' '.join([w for w in jieba.cut(re.sub(r'[^\dA-Za-z\u3007\u4E00-\u9FCB\uE815-\uE864]','', x), 
                               cut_all = False) if w not in stop_words]))
text = ' '.join(df_fs['text_split'].tolist())
generate_wd(text, 
            stop_words, 
            project_dir + '/pic/吃瓜罗志祥分手20200423/luozhixiang_fenshou.png')
```

大众主流的意见就是罗志祥**道歉**。但是其中有些词比较奇怪我们来探索下。

1. 彭昱畅
2. 屈楚　萧黎
3. 蝴蝶

```python
df_fs[df_fs['text_split'] \
    .str.contains('彭昱畅')] \
    .sort_values('num_like', ascending=False) \
    .text
```

罗志祥的分手和彭昱畅的承认恋情放一起是为了产生对比么。。。

```python
df_fs[df_fs['text_split'] \
    .str.contains('屈楚')] \
    .sort_values('num_like', ascending=False) \
    .text
```

大家都把２个瓜放一起吃。。。一个罗志祥一个屈楚萧

```python
df_fs[df_fs['text_split'] \
    .str.contains('蝴蝶')] \
    .sort_values('num_like', ascending=False) \
    .text
```

蝴蝶姐姐，是罗志祥旗下唯一的女艺人。真是一瓜接一瓜，娱乐圈好复杂。。。

```python
df_fs \
    .groupby(['date','hour']) \
    .count() \
    .loc[:, 'id'] \
    .plot(kind = 'bar', 
          figsize=(15, 5),
          title='吃瓜群众几点最活跃',
          color=next(color_cycle))
plt.savefig(project_dir + '/pic/吃瓜罗志祥分手20200423/luozhixiang_fenshou_timeline.png')
plt.show()
```

周扬青第一条事件微博是4月23日 09:00, 到１１点我们的词条#罗志祥周扬青分手#迅速得到关注，之后24号周五上午和下午有大量关注。

```python
df_fs \
    .groupby('platform') \
    .count() \
    .sort_values('id',ascending= False) \
    .head(10) \
    .loc[:, 'id'] \
    .plot(kind = 'bar', 
          figsize=(15, 5),
          title='吃瓜群众爱用什么装备',
          color=next(color_cycle))
plt.savefig(project_dir + '/pic/吃瓜罗志祥分手20200423/luozhixiang_fenshou_platform.png')
plt.show()
```
